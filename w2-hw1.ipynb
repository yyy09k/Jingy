```python
n = int(input())
result = []
if n%3 ==0:
    while n>0:
        result.append(3)
        n-=3
elif n%3==1:
    result.append(2)
    result.append(2)
    n-=4
    while n>0:
        result.append(3)
        n-=3
elif n%3==2:
    result.append(2)
    n-=2
    while n>0:
        result.append(3)
        n-=3
print(result)
```

（1）若n=2001，其正整数列均为3，如上所示。
（2）将一个数字拆分为若干个小数值数字相加，可知数字值越接近时，其乘积越大，故最终均拆分为2或3。
首先判断该数字中含有几个2或3，若数字能被三整除，则数字全为3时拆分出的乘积最大。若数字除三余1，说明该数字含两个2，其他均为3.若数字除三余2，说明该数字有一个2，其他全为3。
